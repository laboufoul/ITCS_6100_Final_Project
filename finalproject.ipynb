Implementing Decision Tree Algorithm to Predicting Closing Price 

```python
# splitting data to be used in all predictive algorithms
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(index_processed.loc[:, ~index_processed.columns.isin(['Close', 'Index', 'Date'])],index_processed['Close'],test_size=0.7,random_state=101)
```

```python
X_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
from matplotlib import pyplot
```

```python
dtclf = DecisionTreeRegressor()

dtclf = dtclf.fit(X_train,y_train)

y_pred = dtclf.predict(X_test)
```

```python
y_pred
```

```python
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
```

```python
pyplot.scatter(y_test, y_pred)
pyplot.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'crimson', lw=4)
pyplot.xlabel('Actual Value')
pyplot.ylabel('Predicted Value')
pyplot.show()
```

Implementing K-Nearest Neighbors Algorithm to Predicting Closing Price 

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X_train, y_train)
```

```python
y_pred = neigh.predict(X_test)
```

```python
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
```

```python
pyplot.scatter(y_test, y_pred)
pyplot.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'crimson', lw=4)
pyplot.xlabel('Actual Value')
pyplot.ylabel('Predicted Value')
pyplot.show()
```

Implementing Random Forest Regressor to Predict Closing Price

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=5, random_state=0)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
```

```python
pyplot.scatter(y_test, y_pred)
pyplot.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'crimson', lw=4)
pyplot.xlabel('Actual Value')
pyplot.ylabel('Predicted Value')
pyplot.show()
```
